**Variables dictionary**

* `id` - Airbnb's unique identifier for the listing

## Cleaning Outliers

To address the outliers of this dataset, we will analyze the statistical distribution, plot boxplots and calculate the limits using the IQR Score rule.

First, let's remember what the IQR is.

The IQR is calculated by subtracting the Third Quartile (75%) from the First Quartile (25%).

# IQR = Q3 - Q1

Let's take a look at our data and see what we identify.

```
# checking the statistical distribution
df_clean.describe().round(1)
```

Here, some things already call our attention, such as:

* The `availability_365` variable has a minimum of 0. Remembering that the price variable deals with the daily price of properties in local currency (EUR), we are seeing that Q3 is at 160 dollars, but the maximum is at 90 thousand dollars. Clearly, there are outliers here.
* The `minimum_nights` variable has a maximum value of 1125, and Q3 is at 34. We clearly have outliers in this variable.
* The variables `number_of_reviews`, `reviews_per_month`, `calculated_host_listings_count`, `availability_365` and `number_of_reviews_ltm` can also contain outliers, but we won't worry about them now.

```
# checking the statistical distribution
df_clean.hist(figsize=(20,15));
```

Checking the histograms, we can clearly see that we have outliers present. To treat them, we will follow these steps:

* Set Q1 and Q3 to the variables that will be cleared.
* Calculate the IQR for the variables.
* Define the upper and lower limit for clipping the outliers.
* Remove outliers.


```
# identifying outliers for the price variable
q1_price = df_clean.price.quantile(.25)
q3_price = df_clean.price.quantile(.75)
IQR_price = q3_price - q1_price
print('IQR of the price variable: ', IQR_price)

# setting the limits
sup_price = q3_price + 1.5 * IQR_price
inf_price = q1_price - 1.5 * IQR_price

print('price upper limit: ', sup_price)
print('lower price limit: ', inf_price)
```

Here we can see that even though we don't have outliers at the bottom, we still have values equal to zero, which need to be dealt with.

Let's plot a boxplot to visualize the difference made by cleaning.

```
# checking the dataset
fig, ax = plt.subplots(figsize=(15,3))
df_clean.price.plot(kind='box', vert=False)
ax.set_title('Dataset - Price')
plt.show()
print("This dataset contains {} columns".format(df_clean.shape[0]))
print("{} inputs over €$299.5".format(len(df_clean[df_clean.price > 299.5])))
print("Meaning {:2f}% of the dataset".format((len(df_clean[df_clean.price > 299.5]) / df_clean.shape[0])*100))
```
